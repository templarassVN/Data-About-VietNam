```python
import pandas as pd
import numpy as np
```

```python
def Fix_northern(data):
    tinh_thieu_CO = np.array(['Cao Bằng','Phú Thọ'])
    tinh_thieu_O3 = np.array(['Thái Nguyên'])
    tinh_thieu_NO2 = np.array(['Quảng Ninh'])
    col_CO = ['CO','NO2','O3','PM-10','PM-2-5','SO2']
    col_O3 = ['O3','PM-10','PM-2-5','SO2']
    col_NO2 = ['NO2','O3','PM-10','PM-2-5','SO2']
    
    fix = data.loc[data['Tỉnh'].isin(tinh_thieu_CO),col_CO]
    done = fix[col_CO].shift(1,axis="columns")
    data.loc[data['Tỉnh'].isin(tinh_thieu_CO),col_CO] = done.values
    
    fix = data.loc[data['Tỉnh'].isin(tinh_thieu_O3),col_O3]
    done = fix[col_O3].shift(1,axis="columns")
    data.loc[data['Tỉnh'].isin(tinh_thieu_O3),col_O3] = done.values
    
    fix = data.loc[data['Tỉnh'].isin(tinh_thieu_NO2),col_NO2]
    done = fix[col_NO2].shift(1,axis="columns")
    data.loc[data['Tỉnh'].isin(tinh_thieu_NO2),col_NO2] = done.values
    
    
    return data
```

```python
data_37 = pd.read_csv("./Crawl Data/aqi_MienBac.txt",sep = '\t')
```

```python
data_37 = Fix_northern(data_37)
tinh_thieu_CO = np.array(['Cao Bằng','Phú Thọ','Quảng Ninh'])
tinh_thieu_O3 = np.array(['Thái nguyên'])
tinh_thieu_NO2 = np.array(['Quảng Ninh'])
data_37[data_37['Tỉnh'].isin(tinh_thieu_CO)]
```

```python
data_nort_add = pd.read_csv("./Crawl Data/Fixed data/aqi_MienBac.txt",sep = '\t')
data_nort_add = Fix_northern(data_nort_add)
```

```python
total = pd.concat([data_37,data_nort_add], ignore_index = True )
```

```python
all_files = ['Bac Ninh.txt','Cao bang.txt','Da nang.txt','Gia lai.txt','Ha noi.txt','Hung Yen.txt','Phu tho.txt','Quang Ninh.txt','Thai nguyen.txt']
for file in all_files:
    path = './Crawl Data/Addition/'
    data_add = pd.read_csv(path + file,sep = '\t')
    data_add = Fix_northern(data_add)
    total = pd.concat([total,data_add],ignore_index = True )

```

```python
def Mien(row):
    bac = np.array(['Cao Bằng', 'Phú Thọ', 'Thái Nguyên',
       'Hà Nội', 'Bắc Ninh', 'Hưng Yên', 'Quảng Ninh'])
    trung = np.array(['Gia Lai', 'Ðà Nẵng','Nghệ An'])
    nam = np.array(['Tp. Hồ Chí Minh','Cần Thơ'])
    if  row['Tỉnh'] in bac:
        return 1
    if row['Tỉnh'] in trung:
        return 2
    if row['Tỉnh'] in nam:
        return 3
final['Miền'] = final.apply (lambda row: Mien(row), axis=1)
```

```python
final.sort_values(by=['Tỉnh','Ngày giờ'])
```

```python
final.drop(final.loc[final['Tỉnh']== 'Hưng Yên'].index, inplace=True)
```

```python
final.to_csv('aqi_VietNam.csv', sep='\t', encoding='utf-8',index = False)
```
